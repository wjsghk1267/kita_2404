```python
import pymysql
from datetime import datetime

# 데이터베이스 연결
connection = pymysql.connect(
    host='localhost',
    user='kevin',
    password='kevin',
    database='kevin_db'
)

cursor = connection.cursor()

# 사용자 데이터
users_data = [
    {
        "username": "user1",
        "email": "user1@example.com",
        "password_hash": "user1",  # 실제로는 비밀번호를 해시하여 저장해야 합니다.
        "is_admin": 0
    },
    {
        "username": "user2",
        "email": "user2@example.com",
        "password_hash": "user2",
        "is_admin": 0
    },
    {
        "username": "user3",
        "email": "user3@example.com",
        "password_hash": "user3",
        "is_admin": 0
    },
    {
        "username": "user4",
        "email": "user4@example.com",
        "password_hash": "user4",
        "is_admin": 0
    },
    {
        "username": "user5",
        "email": "user5@example.com",
        "password_hash": "user5",
        "is_admin": 0
    }
]

# 사용자 데이터 입력
user_insert_query = """
INSERT INTO User (username, email, password_hash, is_admin)
VALUES (%s, %s, %s, %s)
"""
for user_data in users_data:
    cursor.execute(user_insert_query, (user_data["username"], user_data["email"], user_data["password_hash"], user_data["is_admin"]))

# 데이터 커밋하여 사용자 데이터를 저장
connection.commit()

# 사용자 ID 조회
cursor.execute("SELECT id, username FROM User")
user_id_map = {username: user_id for user_id, username in cursor.fetchall()}

# 과제 데이터
tasks_data = [
    {"title": f"Task {i}", "contents": f"Contents of task {i}", "input_date": "2024-07-01", "due_date": "2024-07-10", "user_id": user_id_map[f"user{user_id}"]}
    for user_id in range(1, 6) for i in range((user_id-1)*10+1, user_id*10+1)
]

# 과제 데이터 입력
task_insert_query = """
INSERT INTO Task (title, contents, input_date, due_date, user_id)
VALUES (%s, %s, %s, %s, %s)
"""
for task_data in tasks_data:
    cursor.execute(task_insert_query, (task_data["title"], task_data["contents"], task_data["input_date"], task_data["due_date"], task_data["user_id"]))

# 데이터 커밋하여 과제 데이터를 저장
connection.commit()

# 연결 닫기
cursor.close()
connection.close()
```
