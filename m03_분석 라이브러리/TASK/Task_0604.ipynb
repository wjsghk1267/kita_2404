```python
# Task1. 주어진 df의 날짜 리스트를 datetime 형식으로 변환하고, 변환된 데이터프레임의 년도, 월, 일 열을 추가하시오.
import pandas as pd

# 날짜 문자열 리스트
date_list = ['2023-01-01', '2023-02-15', '2023-03-30']

# 데이터프레임 생성
df = pd.DataFrame({'value': [100, 200, 300]}, index=pd.to_datetime(date_list))

# 인덱스를 기반으로 년, 월, 일 열 추가
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day

print(df)
```

```python
# Task2. 주어진 월간 데이터를 타임스탬프로 변환하고, 변환된 결과에서 시간을 제외한 날짜만 출력하시오.
import pandas as pd

# 문자열을 Period로 변환
period_data = pd.Series(['2023-01', '2023-02', '2023-03'], name='period')
period_data

# 문자열을 Period로 변환
period_data = period_data.astype('period[M]')

# Period를 Timestamp로 변환
timestamp_data = period_data.apply(lambda x: x.to_timestamp())

# 날짜만 출력
print(timestamp_data)
```

```python
# period[M] 타입이 월 단위이기 때문에 to_timestamp()가 해당 월의 첫 날의 날짜로 변환. 각 Period에 시간을 추가하면 시간도 출력
import pandas as pd

now = pd.Timestamp.now()
offset = pd.offsets.MonthBegin()
print(offset.rollforward(now))
# 현재 날짜를 기준으로 가장 가까운 해당 월의 첫날짜로 이동
offset.rollback(now)
```

```python
# Task3. 주어진 날짜 데이터 시리즈를 분기로 변환하고, 각 분기별로 데이터의 개수를 출력하세요.
import pandas as pd

# 날짜 데이터 시리즈
date_series = pd.Series(pd.date_range('2023-01-01', '2023-12-31', freq='Q'))
date_series

# 날짜 데이터를 분기로 변환
quarter_data = date_series.dt.to_period('Q')

# 각 분기별 데이터 개수 세기
quarter_counts = quarter_data.value_counts().sort_index()

print(quarter_counts)
```

```python
# Task5. 2020년부터 2025년까지의 분기별 기간 범위를 생성하고, 각 분기의 시작과 끝 날짜를 포함하는 데이터프레임을 작성하세요.
dates = pd.date_range('2020-01-01','2025-12-31', freq='Q')
start_end = pd.DataFrame(columns = ['Start',' End'], index= range(len(dates)))
for i, Start in enumerate(dates):
    End = Start + pd.offsets.QuarterEnd()
    start_end.loc[i] = [Start, End]

print(start_end)
```

```python
# Task5.teach
period_range = pd.period_range(start='2020Q1',end ='2025Q4', freq='Q')
df = pd.DataFrame({
    'quarter' : period_range,
    'start_date' : period_range.start_time,
    'end_date' : period_range.end_time
})
print(df)
```

```python
# Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.
dates = pd.date_range('2000-01-01','2000-01-10 23:59', freq='12h')
dates
```

```python
# Task6.teach
timedelta_range = pd.timedelta_range(start=0, end='10D', freq='12H')

df = pd.DataFrame({
    'timedelta': timedelta_range,
    'cumulative_time': [timedelta.total_seconds() / 3600 for timedelta in timedelta_range] # 누적 시간을 시간 단위로 계산
})

print(df)
```

```python
# Task8. 주어진 df에서 월별로 데이터의 합계를 계산하세요.
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

#teach
df['month'] = df['date'].dt.to_period('M')
monthly_sum = df.groupby('month')['value'].sum().reset_index()
print(monthly_sum)
```

```python
# Task9. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df

# teach
df['weekday'] = df['date'].dt.dayofweek
weekend_df = df[df['weekday'] >= 5]
print(weekend_df)
```

```python
# Task10. 주어진 df에서 분기별 데이터의 평균을 계산하세요.
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

# teach
df['quarter'] = df['date'].dt.to_period('Q')
quarter_mean = df.groupby('quarter')['value'].mean().reset_index()
print(quarter_mean)
```
