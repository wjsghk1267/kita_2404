```python
# Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.

import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta

# 고객데이터 생성(df1)
random.seed(0)
customer = np.arange(1000)
gender = np.random.randint(0,2,1000)
age = np.random.randint(10,80,1000)
region = np.random.randint(1,10,1000)

c1 = pd.Series(customer, name='customer_id')
c2 = pd.Series(gender, name='gender')
c3 = pd.Series(age, name='age')
c4 = pd.Series(region, name='region')

df1 = pd.concat([c1, c2, c3, c4], axis=1)
df1
```

```python
# 거래데이터 생성(df2)

start_date = datetime(2020, 1, 1)
end_date = datetime(2021, 12, 31)

date = []
for _ in range(10000):
    random_days = random.randint(0, (end_date - start_date).days)
    random_date = start_date + timedelta(days=random_days)
    date.append(random_date)

random.seed(0)
order_no = np.arange(10000)
weekday = [random.choice(['월','화','수','목','금','토','일']) for i in range(10000)]
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
product_20 = np.random.randint(10,30,10000)
product_21 = np.random.randint(10,30,10000)
price_20 = np.random.randint(1000,50000,10000)
price_21 = np.random.randint(1000,50000,10000)
quantity_20 = np.random.randint(1,100,10000)
quantity_21 = np.random.randint(1,100,10000)
time_20 = np.random.randint(1,24,10000)
time_21 = np.random.randint(1,24,10000)
amount_20 = price_20 * quantity_20
amount_21 = price_21 * quantity_21

p1 = pd.Series(order_no, name='order_no')
p2 = pd.Series(weekday, name='weekday')
p3 = pd.Series(date, name='date')
p4 = pd.Series(product_20, name='product_20')
p5 = pd.Series(product_21, name='product_21')
p6 = pd.Series(price_20, name='price_20')
p7 = pd.Series(price_21, name='price_21')
p8 = pd.Series(quantity_20, name='quantity_20')
p9 = pd.Series(quantity_21, name='quantity_21')
p10 = pd.Series(time_20, name='time_20')
p11 = pd.Series(time_21, name='time_21')
p12 = pd.Series(amount_20, name='amount_20')
p13 = pd.Series(amount_21, name='amount_21')

df2 = pd.concat([p1, p2, p3, p4, p5,p6,p7,p8,p9,p10,p11,p12,p13], axis=1)
df2
```

```python
# 통합 데이터프레임 생성(df)
df2['customer_id'] = np.random.randint(1,1000,10000)

df = pd.merge(df1, df2, left_on='customer_id', right_index=True)
```

## 매장운영관리 데이터셋(고객 정보 및 구매이력) 인사이트

1. 기본변수 / 파생변수
 - amount : 제품의 매출액. 파생변수(amount) 매출합계 생성하여 목표변수로 설정. 다른 변수와 상관관계를 분석하여 매장의 다양한 지표를 확인할 수 있음.

 - gender : 고객의 성별. 고객의 성별에 따른 제품의 구매 선호도, 소비액수, 선호 제품의 차이를 비교/분석할 수 있음.

 - age : 고객의 나이. 파생변수(age_group)을 생성하여 연령대에 따라 구매 행동이나 선호도를 비교/분석할 수 있음.

 - region : 고객의 거주지역을 파생변수(region_group)로 그룹화하여 거주지에 따른 에 제품 선호도와 소비 패턴을 분석할 수 있음.

 - product : 각 제품별 매출액을 분석하여 가장 수익성이 높은 제품이 무엇인지 파악할 수 있음. 파생변수 생성(total_product) 및 구분(high,normal,low)하여, 소비자의 소비패턴 변화에 따른 마케팅 전략 개발, 제품의 판매량과 가격에 따라 수익성이 어떻게 변하는지 대응할 수 있음.

 - date : 파생변수(year, month, quarter)등을 생성하여 날짜에 따른 매출 변화로 매장의 성장률, 하락률을 확인할 수 있음. 혹은 특정일(기념일, 이벤트, 프로모션)에 기대 성과 달성률을 확인하는데 활용 가능한 변수.

 - time : 시간대에 따른 매장 이용률 변화를 확인/분석할 수 있음. 시간대별 그룹화하여 (M,A,N)등으로 구분 가능.

 - weekday : 요일별 소비패턴을 분석하여 특정 요일, 특정 제품의 판매량이 높아지는지를 분석할 수 있음. 아마도 실제 매장이라면 주말이나 휴일에는 평일 비해 고객수와 매출이 증가할 것을 예측할 수 있음

```python
# Q2. df를 수정하여 5가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요.

# 파생변수 생성
df_a = df.copy()

# amount 병합,구분 - 매출합계
df_a['amount'] = df_a['amount_20'] + df_a['amount_21']

pd_bins = [0,10,20,30]
pd_labels = ['High', 'Normal', 'Low']
df_a['product_group'] = pd.cut(df_a['product'], bins=pd_bins, labels=pd_labels)

# product 병합,구분 - 제품합계
df_a['product'] = (df_a['product_20'] + df_a['product_21']) / 2

pd_bins = [0,10,20,30]
pd_labels = ['High', 'Normal', 'Low']
df_a['product_group'] = pd.cut(df_a['product'], bins=pd_bins, labels=pd_labels)

# date 구분 (year)
date_bins = pd.to_datetime(['2019-12-31', '2021-01-01', '2022-01-01'])
date_labels = ['2020', '2021']
df_a['Year'] = pd.cut(df_a['date'], bins=date_bins, labels=date_labels)

# 연령대 구분
age_bins = [0,10,20,40,60,80]
age_labels = ['Child', 'Teenager','Adult','Senior', 'Elder']
df_a['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

Baby = df_a['age_group'] == 'Child'
Teen = df_a['age_group'] == 'Teenager'
Adult = df_a['age_group'] == 'Adult
Old = df_a['age_group'].isin(['Senior', 'Elder'])

# 연령대 지출액 변경
df_a.loc[Baby, ['amount']] *= 0.1
df_a.loc[Teen, ['amount']] *= 0.3
df_a.loc[Adult, ['amount']] *= 1.0
df_a.loc[Old, ['amount']] *= 1.5

# 성별 변경
df_a['gender'] = df_a['gender'].map({0: 'Female', 1: 'Male'})

# 요일 변경
weekday_bins = ['월', '화', '수', '목', '금', '토', '일']
weekday_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_a['weekday'] = pd.cut(df_a['weekday'], bins=weekday_bins, labels=weekday_labels)

# 지역 구분
region_bins = [0,3,7,10]
region_labels = ['Up', 'Middle','Down']
df_a['region_group'] = pd.cut(df['region'], bins=region_bins, labels=region_labels)

Up = df_a['region_group'] == 'Up'
Middle = df_a['region_group'] == 'Middle'
Down = df_a['region_group'] == 'Down'
# 지역별 매출액 변경
df_a.loc[Up, ['amount']] *= 2.0
df_a.loc[Middle, ['amount']] *= 1.4
df_a.loc[Down, ['amount']] *= 0.7
```
