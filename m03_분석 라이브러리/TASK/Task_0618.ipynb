```python
# Task4_0618. sex와 class간의 생존율을 보여주는 히트맵을 작성하세요.

t_df = sns.load_dataset('titanic')
t_df.head()
```

```python
t_df['sex'] = t_df['sex'].map({'male':0, 'female':1})
t_df['class'] = t_df['class'].map({'First':1, 'Second':2, 'Third':3})
```

```python
t_df[['sex','class','survived']].corr()
```

```python
sns.heatmap(t_df[['sex','class','survived']].corr(), annot=True)
plt.title('Heatmap of the Survived', pad=20, fontsize=15)
plt.show()
```

```python

# teach
pivot = pd.pivot_table(t_df, index='sex',
                       columns = 'class',
                       values='survived',
                       aggfunc=(lambda x:x.sum()/x.count()), fill_value=0)
sns.heatmap(pivot,annot=True)
```

```python
import seaborn as sns

# Task5_0618. 타이타닉 데이터셋을 사용하여, 객실 등급(class)에 따른 탑승객의 나이를 나타내는 박스 플롯과 바이올린 플롯을 같은 그래프 내에서 나란히 생성하세요.
# 이를 통해 각 클래스별 나이 분포의 차이를 비교하세요.
# 박스플롯
plt.figure(figsize=(8,4))
plt.subplot(1, 2, 1) 
sns.boxplot(x='class', y='age', data=t_df, color ='skyblue')
plt.title('Box Plot')
plt.subplot(1, 2, 2) 
sns.violinplot(x='class', y='age', data=t_df, color='yellow')
plt.title('Violin Plot')
plt.show()
```

```python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>
</body></html>
"""
```

```python
# Task6_0618. 주어진 HTML 문서에서 모든 $<a>$ 태그를 추출하세요.
soup = BeautifulSoup(html_doc, 'html.parser')
soup.find_all('a')

links = soup.find_all('a')
for link in links:
    print(link.text)
```

```python
# Task7_0618. 주어진 HTML 문서에서 첫 번째 $<p>$ 태그의 텍스트를 추출하세요.
soup.find_all('p')

tag = soup.find('p')
print(tag.text)
```

```python
# teach
p_tag = soup.find('p')
print(p_tag.get_text())
```

```python
# Task8_0618. 주어진 HTML 문서에서 모든 링크의 URL을 추출하세요.
links = soup.find_all('a')
for link in links:
    print(link.get('href'))
```

```python
# 주어진 HTML 문서에서 모든 링크의 URL을 추출
a_tags = soup.select('.sister')
for tag in a_tags:
    print(tag['href'])
```

```python
# Task9_0618. 주어진 HTML 문서에서 클래스가 sister인 모든 태그를 추출하세요.
tags = soup.find_all(class_="sister")

for tag in tags:
    print(tag.text)
```
