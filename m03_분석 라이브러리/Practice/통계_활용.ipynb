Q. 어느 도시에 지난 1년간 매달 내린 강수량 데이터는 다음과 같습니다:

[85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95].

이 데이터를 이용해 중간값과 1사분위수, 3사분위수를 계산하여 강수량의 분포를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rain = [85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95]

pd = pd.Series(rain)
m1 = pd.median()
q1 = pd.quantile(0.25)
q3 = pd.quantile(0.75)

print(m1,'\n',q1,'\n',q3)

# 인사이트
quantile_value = [q1, q3]

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.bar(['Q1','Q3'], quantile_value, color=['pink', 'skyblue'])
plt.title('Precipitation')
plt.xlabel('Quantile')
plt.ylabel('Rain')

# teach insight
# 중간값은 전체 강수량 분포의 중심을 나타내고, 사분위수는 데이터의 분포 범위를 나타낸다.
print(f'강수량의 중앙값은 {m1}mm이며, 이는 일반적인 강수량 수준을 나타낸다.')
print(f'1사분위수와 3사분위수의 차이는 {q3-q1}mm이며, 이는 강수량의 변동성을 보여준다')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

salaries = [45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000]

plt.boxplot(salaries)
plt.title('Salay')
plt.ylabel('salary')
plt.show()

#인사이트 해석
median_salary = np.median(salaries)
q1_salary = np.percentile(salaries, 25)
q3_salary = np.percentile(salaries, 75)
iqr_salary = q3_salary - q1_salary
std_dev_salary = np.std(salaries)
mean_salary = np.mean(salaries)

print(f'Median: {median_salary}')
print(f'Q1: {q1_salary}, Q3: {q3_salary}, IQR: {iqr_salary}')

# 이상치 확인 - 없음.
lower_bound = q1_salary - 1.5 * iqr_salary
upper_bound = q3_salary + 1.5 * iqr_salary
outliers = [salary for salary in salaries if salary < lower_bound or salary > upper_bound]

print(f'Outliers: {outliers}')
```

Q. 한 마케팅 캠페인에서 투자한 광고비와 발생한 매출 데이터가 각각 다음과 같습니다:

광고비 [1, 2, 3, 4, 5] (단위: 천 달러),

매출 [10, 15, 20, 25, 30] (단위: 천 달러).

광고비와 매출 간의 관계를 시각화하기 위해 산점도를 그리고, 두 변수 간의 관계에 대한 인사이트를 도출하세요.

```python
value = [1, 2, 3, 4, 5]
profit = [10, 15, 20, 25, 30] + np.random.randint(10, size=5)

plt.scatter(value, profit)
plt.title('marketing')
plt.show()

# 인사이트 해석
correlation = np.corrcoef(value, profit)[0,1]
print(f'Correlation: {correlation}')

if correlation > 0.7:
    print('광고비 : 매출은 양의 상관관계이다')

else :
    print('광고비 : 매출은 상관관계가 약하거나 없다')
```

Q. 한 식당의 월별 홍보 비용과 월별 방문자 수 데이터가 각각 다음과 같습니다:

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
import pandas as pd
visitor = [300, 400, 500, 600, 700]
cost = [500, 600, 700, 800, 900]

g1 = pd.Series(visitor)
g2 = pd.Series(cost)
correlation = g1.corr(g2)

print('상관계수:',round(g1.corr(g2)))


# 인사이트 해석
if correlation > 0.7:
    print('홍보비용 : 방문자 수는 양의 상관관계이다')

elif correlation < 0.7:
    print('홍보비용 : 방문자 수는 음의 상관관계이다')

else :
    print('홍보비용 : 방문자 수는 상관관계가 약하거나 없다')
```

```python
v_count = np.sort(visitor)[::-1] # 역순정렬

# 상관계수 계산
correlation = np.corrcoef(cost, v_count)[0,1]
print(f'Correlation: {correlation}')

# 인사이트 해석
if correlation > 0.7:
    print('홍보비용 : 방문자 수는 양의 상관관계이다')

elif correlation < 0.7:
    print('홍보비용 : 방문자 수는 음의 상관관계이다')

else :
    print('홍보비용 : 방문자 수는 상관관계가 약하거나 없다')
```

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cost = [100, 200, 300, 400, 500]
profit = [200, 400, 600, 800, 1000]

df = pd.DataFrame(cost, columns=['x'])
df['y'] = profit

plt.scatter(cost, profit)
m, b = np.polyfit(df['x'], df['y'], 1)
plt.plot(df['x'], m*df['x'] + b, color='red', label='Trend line')
plt.title('R&D')
plt.show()
```
