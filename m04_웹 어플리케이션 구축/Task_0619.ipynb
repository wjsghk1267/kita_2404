```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.naver.com'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')


# Task1_0619. 다음 사항을 수행하세요.
# - 첫 번째로 매칭되는 'p' 태그 찾기
# - 모든 'p' 태그 찾기
# - 클래스가 'content'인 첫 번째 'p' 태그 찾기
# - 클래스가 'content'인 모든 'p' 태그 찾기

tag_p = soup.find('p')
sel_p = soup.select_one('p')
tag_p_all = soup.find_all('p')
sel_p_all = soup.select('p')

class_p = soup.find(class_='content')
class_p_all = soup.find_all(class_='content')


print('첫번째 p tag:',tag_p,'\n',sel_p)
print('모든 p tag:',tag_p_all,'\n',sel_p_all)
print('클래스가 content인 첫번째 p태그:',class_p)
```

```python
# - 특정 'p' 태그의 모든 부모 태그 찾기
# - 특정 'p' 태그의 첫 번째 부모 태그 찾기
# - 특정 'p' 태그의 다음 형제 태그 찾기
# - 특정 'p' 태그의 이전 형제 태그 찾기

for parent in tag_p.parents:
    print(parent.name)


parent_f = tag_p.parent()
pre_siblings = tag_p.find_previous_sibling()
next_siblings = tag_p.find_next_sibling()

print(parent_f)
print(pre_siblings)
print(next_siblings)
```

```python
p_tags = soup.find_all('p')

# - 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
for p_tag in p_tags:
    siblings = p_tag.find_next_siblings()
    for sibling in siblings:
        print('다음:',sibling.name, sibling.get_text(strip=True))

# - 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 
for p_tag in p_tags:
    siblings = p_tag.find_previous_siblings()
    for sibling in siblings:
        print('이전:',sibling.name, sibling.get_text(strip=True))
```

```python
# Task2_0619. ID를 이용해서 '네이버 뉴스' 추출하세요.
import requests
from bs4 import BeautifulSoup
url = 'https://news.naver.com'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 동적 콘텐츠 로딩 : selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

options = Options()
options.add_argument("--start-maximized")
options.add_experimental_option("detach",True)

driver = webdriver.Chrome(options=options)

# 네이버 페이지 열기
url = 'https://news.naver.com/'
driver.get(url)

# 페이지 로딩 대기
time.sleep(2) # 5초 대기

# 뉴스스탠드 링크 찾기 : #newsstand > div.ContentHeaderView-module__content_header___nSgPg > div > ul > li:nth-child(1) > span > a:nth-child(1)
try:
    newsstand_link = driver.find_element(By.XPATH, '/html/body/section/header/div[1]/div/div/div[1]/div/h1')
    print(newsstand_link.text)
except:
    print('링크없음')

driver.quit()
```
