객체 탐지(Object Detection)는 이미지나 비디오에서 객체의 위치와 종류를 인식하는 컴퓨터 비전 기술입니다. 이 기술은 딥러닝의 발전과 함께 급격히 발전해왔으며, 주로 자율주행, 영상 분석, 보안, 의료 등 다양한 분야에 응용됩니다. 객체 탐지 모델의 발전 과정은 크게 세 가지로 나눌 수 있습니다:

### 1. **전통적 방법 (Traditional Methods)**

딥러닝 이전의 객체 탐지는 주로 기계 학습과 수동으로 설계된 특징 추출기(Feature Extractor)를 활용했습니다. 이 방법들은 주로 경계 상자(bounding box)를 직접 탐지하기보다는 이미지의 주요 특징을 추출한 후 분류기를 사용해 객체를 인식했습니다. 그러나 이 전통적인 방식은 계산 비용이 높고, 다양한 상황에서 일관된 성능을 내기 어렵다는 한계가 있었습니다. 이에 따라 딥러닝 기반 방법들이 등장했습니다.

### 2. **초기 딥러닝 모델 (Early Deep Learning Models)**

딥러닝의 발전과 함께, 전통적인 특징 추출기 대신 CNN(Convolutional Neural Network)을 활용한 모델들이 등장했습니다. 특히 **R-CNN(Region-based Convolutional Neural Network)** 계열 모델들이 발전의 핵심이었습니다.

- **R-CNN (2014)**: 이미지를 여러 개의 영역으로 분할한 후, 각 영역을 CNN을 통해 처리해 객체를 탐지합니다. 이 모델은 뛰어난 성능을 보였지만, 매우 느렸고 많은 계산 자원이 필요했습니다.
- **Fast R-CNN (2015)**: R-CNN의 개선된 버전으로, 영역 제안(region proposals)과 CNN 처리 과정을 병렬화하여 속도를 크게 향상시켰습니다.
- **Faster R-CNN (2016)**: R-CNN 계열 중 가장 발전된 형태로, 이미지에서 영역을 제안하는 과정 자체를 CNN 구조 안에 포함시켜 객체 탐지 속도를 크게 개선했습니다.

### 3. **단일 단계 객체 탐지 (Single-Stage Object Detectors)**

R-CNN 계열 모델은 **두 단계(two-stage)** 프로세스를 사용했지만, 이를 단순화한 **단일 단계(single-stage)** 객체 탐지 모델들이 등장하며 속도와 정확도 면에서 큰 발전을 이루었습니다.

- **YOLO (You Only Look Once, 2016)**: YOLO는 이미지를 그리드(grid)로 나누고 각 그리드에서 객체를 한 번에 예측하는 방식으로 매우 빠른 속도를 자랑합니다. 다만 초기 버전에서는 작은 객체에 대한 탐지 성능이 낮았습니다.
- **SSD (Single Shot Multibox Detector, 2016)**: SSD는 여러 크기의 경계 상자를 한 번에 예측하며, 다양한 크기의 객체를 탐지할 수 있도록 설계되었습니다. YOLO와 비슷한 속도를 유지하면서도 높은 정확도를 보였습니다.

이 모델들은 실시간 처리에 적합하며, 자율주행차와 같은 실시간 객체 탐지가 필요한 곳에서 주로 사용됩니다.

### 4. **최신 객체 탐지 모델 (Current State-of-the-Art Models)**

최근의 객체 탐지 모델들은 정확도와 속도 모두를 향상시키기 위한 다양한 기술들이 도입되었습니다.

- **YOLOv4/v5 (2020)**: YOLO는 지속적으로 개선되었으며, YOLOv4와 v5는 더 높은 성능과 효율성을 목표로 다양한 최적화 기법을 도입했습니다. 특히 YOLOv5는 경량화되어 모바일 및 임베디드 장치에서도 효율적으로 동작할 수 있습니다.
- **EfficientDet (2020)**: EfficientDet은 효율성을 극대화하기 위해, EfficientNet 아키텍처와 비슷한 아이디어를 사용한 모델입니다. 정확도와 속도의 균형을 잘 맞춰 다양한 애플리케이션에서 사용 가능합니다.
- **DETR (Detection Transformer, 2020)**: 딥러닝과 Transformer 구조를 결합한 모델로, 객체 탐지에 Transformer를 처음 적용한 사례입니다. 기존 CNN 기반 방식들과 달리, 별도의 경계 상자 제안 과정 없이 직접적으로 객체를 탐지할 수 있어 구조적으로 더 간단합니다.

현재의 객체 탐지 기술은 정확성과 효율성을 모두 충족시키며, 많은 산업에서 중요한 역할을 하고 있습니다.


## 객체 탐지(Object Detection)
- 이미지나 비디오 내에서 객체의 존재와 위치를 파악하고, 그 객체가 어떤 클래스(예: 사람, 자동차, 동물 등)에 속하는지 식별하는 컴퓨터 비전 기술입니다.
- 객체 탐지는 객체를 분류하는 것(Classification)과 함께 해당 객체의 위치를 사각형으로 표시하는 것(Localization)을 포함합니다.
- 이 기술은 보안 감시, 자율 주행 차량, 이미지 검색 개선, 건강 관리, 로봇 시각 시스템 등 다양한 분야에서 활용됩니다.

[ 기본 개념 ]
- 분류(Classification): 이미지에 특정 객체가 포함되어 있는지 여부를 판별합니다. 예를 들어, 이미지에 고양이가 있는지 없는지를 결정합니다.
- 위치 식별(Localization): 하나의 객체가 이미지 내에 있을 때, 그 객체의 위치를 지정합니다. 일반적으로 사각형 바운딩 박스(Bounding Box)를 사용하여 위치를 나타냅니다.
- 객체 탐지(Object Detection): 이미지 내의 여러 객체를 식별하고, 각 객체의 위치를 사각형 바운딩 박스로 표시합니다.

## YOLO (You Only Look Once)
- YOLO는 객체 감지를 위한 가장 유명한 신경망 아키텍처 중 하나입니다. 이름에서 알 수 있듯이, YOLO는 이미지를 한 번만 보고 객체를 감지할 수 있습니다. 이는 YOLO가 슬라이딩 윈도우와 같은 방법을 사용하지 않고, 전체 이미지를 하나의 큰 그리드로 나누어 각 셀이 객체의 중심을 포함하면 그 객체를 감지하는 방식을 사용하기 때문입니다. YOLO는 이미지를 그리드로 나누고, 각 그리드 셀에서 바운딩 박스와 해당 바운딩 박스에 대한 클래스 확률을 예측합니다. 이러한 특징 때문에 YOLO는 빠른 처리 속도와 높은 정확도를 동시에 제공하며, 실시간 객체 감지에 매우 적합합니다.

- YOLOv5는 YOLOv4에 비해 크기와 속도를 향상시키는데 중점을 두었습니다. 또한, YOLOv5는 사용하기 쉬운 사용자 경험을 제공하며, 모델 학습과 추론 속도가 빠르고, 모델 크기가 작습니다.

- 예제 및 사용 사례
객체 감지는 많은 실제 애플리케이션에서 사용됩니다. 예를 들어, 자율주행 자동차는 객체 감지를 사용하여 주변의 차량, 보행자, 사물 등을 감지하고, 이를 바탕으로 주행 경로를 계획합니다. 또한, 보안 카메라는 객체 감지를 사용하여 움직이는 사람이나 사물을 감지하고, 이상 행동을 식별합니다. 또한, 소셜 미디어 애플리케이션은 객체 감지를 사용하여 사용자가 업로드한 이미지에서 사람이나 사물을 자동으로 태그합니다.

mAP 정리, Precision과 Recall의 이해 : https://lapina.tistory.com/98
- AP (Average Precision) : Recall value [0.0, 0.1, …, 1.0] 값들에 대응하는 Precision 값들의 Average 이다.
- mAP (mean Average Precision) : 1개의 object당 1개의 AP 값을 구하고, 여러 object-detector 에 대해서 mean 값을 구한 것이 mAP 이다.
- 즉 mAP란 mutiple object detection 알고리즘에 대한 성능을 1개의 scalar value로 표현한 것이다.

mAP@0.5 vs mAP@0.5-0.95
- IOU를 0.5로 설정하여 모델 성능을 평가할 경우 원래 Ground true Box와 절반 이상만 겹치는 예측을 하면 맞는 걸로 평가합니다. 즉 절반만 겹치던, 완전히 동일하게 겹치던 모두 다 맞는 걸로 인정하기 때문에 상대적으로 성능이 떨어지는 모델의 평가가 제대로 평가되지 않을 수 있다.
- 예전에 Pascal VOC Contest할 때는 이 정도 수준에서 맞는 걸로 평가했습니다만, Coco Contest에서는 평가 기준을 바꿔 버린다.
- IOU가 0.5 부터 시작해서 보다 엄격한 기준을 적용해 가면서 평가한다.
- IOU가 0.95 정도 되면 거의 Ground Trueth Box와 거의 동일한 수준으로 예측해야 합니다. 이 경우는 상대적으로 IOU가 0.5 일때 보다 모델 성능 평가가 떨어지게 된다.
- 이처럼 IOU를 0.5 부터 조금씩 증가 시켜서 모델 성능 평가를 하는 방식으로 보다 엄격한 모델 성능 평가가 가능

## YOLO (You Only Look Once)
실시간 객체 탐지를 위한 방법론으로, 빠른 처리 속도와 높은 성능을 제공합니다. YOLO는 여러 버전이 있으며, 각 버전은 성능 향상, 속도 개선, 그리고 새로운 기능 추가를 목표로 개발되었습니다. 다음은 YOLOv3, YOLOv4, YOLOv5, 그리고 YOLOv8의 주요 특징과 차이점을 설명한 것입니다.

- YOLOv3:
  - YOLO의 세 번째 버전으로, 성능 향상과 다양한 크기의 객체 탐지를 목표로 하며
  3개의 다른 크기의 피쳐 맵을 사용하여 객체를 탐지.
  - 이를 통해 다양한 크기의 객체를 더 효과적으로 탐지할 수 있다.
  - YOLOv3는 세 가지 다른 크기의 앵커 박스를 사용하여 다양한 모양의 객체를 탐지하는 데 효과적.
  - 세 개의 IoU 임계값을 사용하여 더 많은 양의 예측을 수행.
- YOLOv4:
  - YOLOv3의 성능을 더욱 향상시키기 위해 개발.
  - CSPDarknet53라는 새로운 backbone 네트워크, PANet과 SAM block을 도입하였으며 이들은 피쳐 피라미드 네트워크를 강화하고, 객체 탐지의 정확도를 높인다.
  - Mish 활성화 함수와 CIOU loss 등의 최신 기법을 도입하여 성능을 더욱 향상시켰다.
- YOLOv5:
  - 속도와 사용 편의성을 강조.
  - YOLOv4의 구조를 기반으로 하며, 더 빠른 추론 속도를 제공.
  - 모델의 크기를 조정할 수 있는 여러 버전을 제공하여, 다양한 환경과 요구 사항에 맞게 사용할 수 있다.

- YOLOv8 :
  - 2023년 1월 발표된 버전으로, YOLO 모델을 위한 완전히 새로운 리포지토리를 출시하여 개체 감지, 인스턴스 세분화 및 이미지 분류 모델을 train하기 위한 통합 프레임워크 로 구축됨

## Yolov5

You Only Look Once 의 약자로 Object detection 분야에서 많이 알려진 모델이다.

처음으로 one-stage-detection 방법을 고안하여 실시간으로 Object Detection이 가능하게 만들었다.

#### Yolov5 특징
1. 이미지 전체를 한번만 보는 것

R-CNN: 이미지를 여러장으로 분할하고 CNN 모델을 이용하여 이미지를 분석했다. 따라서 이미지 한장을 보더라도 여러장의 이미지를 분석하는 것과 같았다. 하지만 YOLO는 이러한 과정없이 이미지를 한 번만 보는 특징을 가지고 있다.

### Object Detection의 구조와 종류 : https://ssong915.tistory.com/43

- 백본(Backbone)은 입력이미지를 feature map으로 변형시켜주는 부분이다.

- 헤드(Head)는 Backbone에서 추출한 feature map의 location 작업을 해주는 부분으로, predict classes 와 bounding boxes 작업이 수행된다. 헤드는 크게 Dense Prediction과 Sparse Prediction으로 나누어진다.

- Sparse Prediction을 사용하는 Two-Stage Detector: Predict Classes 와 Bounding Box Regression 분리 ex) R-CNN, R-FCN
- Dense Prediction을 사용하는 One-Stage Detector: Predict Classes 와 Bounding Box Regression 통합 ex) YOLO, SSD

- 넥(Neck) 은 Backbone 과 Head를 연결하는 부분으로, feature map을 정제하고 재구성한다.

### 절차
input(이미지) → CNN → FC(Fully Connected) → PT(Prediction Tensor) 를 반복
- 이미지를 S*S의 그리드로 분할 (논문에서는 7*7)
그리드마다 하나의 객체를 예측한다. 미리 설정해 둔 경계를 통해서 객체의 위치와 크기를 파악한다. 여러 객체가 한 셀에 있으면 탐지를 못할 수도 있다. 각 Grid cell 은 조건이 있다.
 - B개의 boundary boxes를 예측하고, 각 box는 하나의 box confidence score을 가지고 있다.
    - 예측된 box 수에 관계 없이 단 하나의 객체만 탐지한다.
    - boundary box는 객체의 위치 (x,y), 객체의 크기 (w,h), box confidence score로 구성된다.
    - box confidence score는 box가 객체를 포함하고 있을 가능성을 반영한다.
 - C개의 conditional class probabilities를 예측한다.
    - 탐지된 객체가 어느 특정 클래스에 속하는 지에 대한 확률
- 이미지 전체를 신경망에 넣고 특징 추출을 통해 예측 텐서(Prediction Tensor) 생성
    - S * S * (B*5 +C)
    - 7x7 grids, 2 boundary boxes, 20 classes = 7* 7 * 30
    - 이제 YOLO는 7730 텐서를 예측하는 CNN 네트워크를 구축해야한다.



- 이미지 분류는 GT가 이미지의 클래스이기 때문에 이미지가 딱 주어졌을 때 이 이미지가 GT의 클래스와 일치하는지 불일치하는지를 비교
- 객체 검출에서는 이미지를 분류하는 문제와 달리 여러 객체들이 존재하는 이미지 내에서 객체가 어디에 위치하는지 바운딩 박스(Bounding box)로 찾고, 찾은 위치에 해당하는 박스 내부의 객체가 실제 GT에 있는 클래스와 일치하는지의 여부를 비교
- 객체 검출은 클래스를 제대로 예측하였는지 부터 판단하여 예측하지 못하면 실패로 간주한다. 클래스를 올바르게 예측하였다면 바운딩 박스의 정확도를 기준으로 정확도를 측정

객체 검출에서는 GT 와 예측 값 간의 교집합과 합집합을 통해 IoU를 측정
- True Positive (TP) : 맞다고 추측하고 실제로 맞음. IOU >= Threshold
- False Positive (FP) : 맞다고 추측하고 실제로는 틀림. IOU < Threshold
- False Negative (FN) : 아니라고 추측하고 실제로는 맞음. Ground Truth를 아예 detect 못함
- True Negative (TN) : 아니라고 추측하고 실제로 아님.


여기서 Threshold 는 일반적으로 50%, 75%, 95%로 설정된

YOLO 객체 탐지에서 앵커 박스, 바운딩 박스, 그리고 NMS는 중요한 역할을 수행

- 앵커 박스 (Anchor Box):
  - 객체 탐지 알고리즘이 객체를 탐지하기 위한 참조점으로 사용하는 박스.
  - 일반적으로 여러 가지 다른 크기와 비율로 설정되며 알고리즘이 다양한 크기와 비율의 객체를 탐지할 수 있도록 돕는다.
  - 객체 탐지 모델은 각 앵커 박스가 주어진 객체를 포함할 확률을 예측하고, 앵커 박스가 어떻게 조정되어야 해당 객체를 정확하게 둘러싸게 될지를 학습하는 과정을 통해 앵커 박스는 최종적으로 예측된 바운딩 박스로 변환
  - 앵커 박스 없이 객체를 직접 예측하려면 많은 가능한 위치, 크기, 비율을 고려해야 하므로 학습이 어렵고 비효율적. 앵커 박스를 사용하면 모델은 각 앵커 박스에 대한 객체의 존재 여부와 앵커 박스의 조정 방향만을 예측하면 되므로 학습이 더욱 효율적이고 정확

- 바운딩 박스 (Bounding Box):
  - 바운딩 박스는 객체 탐지 알고리즘이 이미지 내에서 객체의 위치를 표시하는 데 사용하는 사각형 박스이며 일반적으로 모델이 객체의 위치를 예측한 결과로, 객체의 중심 좌표, 그리고 박스의 너비와 높이를 나타낸다.

- NMS (Non-Maximum Suppression):
  - 여러 개의 바운딩 박스 중 가장 확신도가 높은 바운딩 박스를 선택하고, 그 바운딩 박스와 많은 부분이 겹치는 다른 바운딩 박스를 제거하는 과정.
  - 이는 객체를 여러 번 탐지하는 것을 방지하고, 객체를 둘러싼 바운딩 박스의 수를 줄여 처리 효율성을 높인다.
  - NMS는 겹치는 바운딩 박스 중에서 IoU (Intersection over Union)라는 지표를 기반으로 가장 높은 점수를 가진 바운딩 박스를 선택.

이 세 가지 개념은 객체 탐지에서 서로 상호 작용하며, 모델이 이미지 내의 객체를 정확하게 탐지하고 위치를 표시할 수 있게 돕는다.

#### 매직 커멘드
- IPython은 Python 쉘을 확장하여 인터랙티브한 컴퓨팅 환경을 제공하는 도구이며 매직 커맨드라는 특수 명령어를 제공하는데, 이는 일반적인 Python 코드와는 달리 특별한 기능을 수행한다. 매직 커맨드는 % 문자로 시작하는 라인 매직과 %% 문자로 시작하는 셀 매직이 있다.

- from IPython.core.magic import register_line_cell_magic는 사용자 정의 매직 커맨드를 만들기 위한 데코레이터를 가져오는 구문.

  - register_line_magic 데코레이터를 사용하여 라인 매직 커맨드를 정의할 수 있으며 이 커맨드는 한 줄에 적용.
  - register_cell_magic 데코레이터를 사용하여 셀 매직 커맨드를 정의할 수 있으며 이 커맨드는 전체 셀에 적용.

```python
# register_line_cell_magic 데코레이터를 이용하여 커스텀 매직 함수인 writetemplate을 정의
# line : 파일 이름으로 사용 될 문자열입니다
# cell : 쓰여질 파일의 내용으로 사용 될 문자열입니다

from IPython.core.magic import register_line_cell_magic

@register_line_cell_magic
def writetemplate(line, cell):
    with open(line, 'w') as f:
        f.write(cell.format(**globals()))
```
