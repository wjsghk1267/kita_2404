1. 명확하고 구체적인 언어 사용

모호한 표현 대신 구체적 단어 사용: 일반적 표현보다 구체적인 단어를 사용하여 명확히 전달합니다. 예: "빠르다" 대신 "5분 내에".
명확한 문장 구조 유지: 단문을 사용하여 문장이 복잡하지 않도록 합니다.

```python
prompts = [
    "2024년도 상반기 경제에 대한 분석",
    "2024년도 하반기 경제에 대한 예측",
    "재무제표 각 평가 수치에 대한 기준 설명",
    "2024년도 엔비디아의 재무제표 가치 평가"
]

for prompt in prompts:
    completion = client.chat.completions.create(
    model='gpt-4o-mini-2024-07-18',
    messages=[
        {'role': 'system', 'content': '당신은 경제 전문가이다'},
        {'role': 'user', 'content': prompt},
    ]
)
    print(f'프롬프트: {prompt}')
    print(completion.choices[0].message.content)
    print('\n'+'='*100 + '\n')
```

```python
prompts = [
    """지시사항: "이번 달 마케팅 성과를 분석하세요."

    프로젝트 개요: "이 프로젝트는 새로운 제품 라인의 시장 진입을 위한 마케팅 캠페인을 진행하는 것입니다. 이번 달 캠페인은 소셜 미디어, 이메일 마케팅, 그리고 협찬 이벤트로 구성되었습니다."
    검토 목적: "이 검토의 목표는 각 마케팅 채널의 성과를 평가하여, 향후 캠페인 전략을 최적화하기 위한 인사이트를 도출하는 것입니다."
    """
]

for prompt in prompts:
    completion = client.chat.completions.create(
        model='gpt-4o-mini-2024-07-18',
        messages=[
            {'role': 'system', 'content': '당신은 마케팅 분석 전문가이다'},
            {'role': 'user', 'content': prompt},
        ]
    )
    print(f'프롬프트: {prompt}')
    print(completion.choices[0].message.content)
    print('\n' + '=' * 100 + '\n')
```
