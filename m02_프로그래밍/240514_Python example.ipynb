#### 2. 아래 list_input에 대해서 주어진 사용자 함수를 일괄적으로 적용할 수 있는 2개의 내장함수를 활용하여 [1, 4, 9, 16, 25], [1, 2]를 출력하는 코드를 작성하세요.

```python
# 사용자 함수를 선언합니다.
def power(item):
    return item * item
def under_3(item):
    return item < 3

# 변수를 선언합니다.
list_input = [1, 2, 3, 4, 5]

result1 = []
for i in list_input:
    item = power(i)
    result1.append(item)
print(result1)

result2 = []
for i in list_input:
    if under_3(i):
        result2.append(i)
print(result2)
```

- 4 입력시 \
정수 입력> 4 \
4번째 요소: 100
예외 처리 완료

- 문자 입력시 \
정수 입력> 가 \
정수를 입력해 주세요!
예외 처리 완료

- 5 입력시 \
정수 입력> 5 \
리스트의 인덱스를 벗어났어요!
예외 처리 완료

```python
list_number = [52, 273, 32, 72, 100]
while 1:
    try:
        user_input = int(input('정수 입력(종료:999)>'))
        if user_input == 999:
            break
        if user_input < 5:
            value = list_number[user_input]
            print(f'{user_input}번째 요소: {value}')

    except IndexError:
        print('리스트의 인덱스를 벗어났어요!')
    except ValueError:
        print('정수를 입력해 주세요!')
    finally:
        print('예외 처리 완료')
```

####  16. numbers 내부에 들어 있는 숫자가 몇 번 등장하는지를 출력하는 코드를 작성하세요.

#### 17. 클래스 Country를 완성하여 아래와 같이 출력하세요.

```python
class Country:
    def __init__(self, name, pop, cap):
        self.name = name
        self.pop = pop
        self.cap = cap
    def desc(self):
        print(f'{self.name}의 인구는 {self.pop}천만명이며 수도는 {self.cap}입니다')

class Korea(Country):
    def desc1(self,gnp):
        self.gnp = gnp
        print('1인당 gnp는 USD %d만입니다.' %self.gnp)

obj = Korea('한국',5,'서울')
obj.desc()
obj.desc1(3)
```

#### 18. 사칙연산을 수행하는  클래스를 작성하여 객체를 만들고 2개의 정수 10, 5를 입력한 후 연산 결과를  포멧을 이용하여 출력하세요.

```python
class CalF():
    def Sum(self,a,b):
        return a+b
    def Sub(self,a,b):
        return a-b
    def Mul(self,a,b):
        return a*b
    def Div(self,a,b):
        return a/b

obj1 = CalF()
result1 = obj1.Sum(10,5)
result2 = obj1.Sub(10,5)
result3 = obj1.Mul(10,5)
result4 = obj1.Div(10,5)

print(f'더하기: {result1}\n빼기: {result2}\n곱하기: {result3}\n나누기: {result4}')
```

#### 20. 문자열 리스트 states를 아래 [출력]와 같이 전처리하여 출력하세요
`states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']`

[출력]\
['Alabama',\
 'Georgia',\
 'Georgia',\
 'Georgia',\
 'Florida',\
 'South Carolina',\
 'West Virginia']

```python
import re

states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south carolina##', 'West virginia?']

def clean_string(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]','',value)
        value = value.title()
        result.append(value)
    return result
clean_string(states)
```
