#### 조건문 (Conditional Statements):
- if: 특정 조건이 참(True)일 때 코드 블록을 실행
    - 조건은 비교 연산자(==, !=, <, >, <=, >=)와 논리 연산자(and, or, not)를 사용하여 표현할 수 있다
- elif: 이전의 if 또는 elif 조건이 거짓(False)일 때 다른 조건을 검사
    - elif는 'else if'의 축약형으로, 이전 if 또는 elif 문의 조건이 거짓(False)일 때 다른 조건을 검사
    - 여러 elif 문을 사용하여 다양한 조건을 검사할 수 있다
- else:
    - 이전의 if 또는 elif 문의 조건이 모두 거짓일 때 실행할 코드 블록을 지정

```python
x = 10
if x > 5:
    print("X는 5보다 크다")
elif x == 5:
    print("X는 5이다")
else:
    print("X는 5보다 작다")


x = 10
if x >= 5:
    print('x는 5보다 크거나 같다')
else:
    print('x는 5보다 작다.')

x = 10
y = 5

if x > 5:
    print("x는 5보다 크다")
if y < 10:
    print("y는 10보다 작다")
if x > y:
    print('x는 y보다 크다')


if x > 5:
    if y >= 3:
        print('x는 5보다 크고 y는 3보다 크거나 같다')
    else:
        print('x는 5보다 크고 y는 3보다 작다')
else:
    print('x는 5보다 작거나 같다')


```

#### 반복문 (Loop Statements):
- for: 시퀀스(리스트, 튜플, 문자열 등)의 아이템을 순회하면서 코드 블록을 실행
- while: 특정 조건이 참인 동안 코드 블록을 반복하여 실행

```python
# for 루프는 시퀀스(리스트, 튜플, 문자열 등)나 반복 가능한 객체(iterable)의 요소를 순회

for i in range(5):
    print(i,end= ' ')
print('\n')
count = 0
while count < 5:
    print(count, end = ' ')
    count += 1
print('\n')
# 리스트를 순회하는 예
for i in [1,2,3,4,5]:
    print(i, end= '\n')

# 문자열을 순회하는 예
for char in 'hello':
    print(char, end =',')
print('\n')

# while 루프는 주어진 조건이 True 동안 코드 블록을 반복
# 조건이 거짓이 되면 루프 종료.
count = 0
sum = 0
while count < 5:
    sum += count
    print(sum, end = ' ')
    count += 1
print('\n')
# Q. d에서 반복문을 사용하여 97 98 99 을 출력하세요.
d = {'a':97, 'b':98, 'c':99}

for key in d:
    print(d[key], end = ' ')
```

dict_values 클래스의 이해:
- 클래스 정의: 파이썬에서 클래스는 객체를 생성하는 데 사용되는 템플릿. dict_values는 딕셔너리의 값을 보여주는 뷰 객체를 생성하는 특별한 내장 클래스.
- 뷰 객체: dict_values 객체는 딕셔너리의 값에 대한 동적인 뷰를 제공. "뷰(view)"란 원본 딕셔너리 객체의 값이 변하면 그 변화가 dict_values 객체에도 반영되는, 원본 데이터의 실시간 반영본을 의미. 이는 딕셔너리의 값들이 수정, 추가, 삭제될 때 dict_values 객체가 자동으로 업데이트된다는 의미.
- 객체 인스턴스: dict_values 객체는 dict_values 클래스의 인스턴스. 이 클래스는 파이썬의 다른 모든 클래스처럼 특정 데이터 구조와 메서드를 가진다. 예를 들어, dict_values 객체는 순회 가능하며 (iterable), 딕셔너리의 모든 값을 순차적으로 접근할 수 있는 메서드를 제공.

#### 제어문 관련 키워드:
- break: 현재의 반복문을 중단하고, 반복문 바깥의 코드를 실행
- continue: 반복문의 나머지 부분을 건너뛰고, 다음 반복을 시작

```python
for i in range(5):
    if i == 3:
        break
    print(i,end= ' ')

print('\n')

for i in range(5):
    if i == 3:
        continue
    print(i,end= ' ' )

print('\n')

for i in range(10):
    if 1 == 5:
        break # i가 5일 때, 루프 종료
    print(i, end = ' ')

print('\n')

for i in range(10):
    if i % 2 == 0:
        continue # i가 짝수일 때, 건너뛰고 다음 반복 실행
    print(i, end = ' ')
print('\n')
for i in range(10):
    if i %2 == 0:
        print(i, end = ' ')

print('\n')

# for 또는 while 루프에 else절을 사용하면, 루프가 정상적으로 완료된 후에 실행한다.
for i in range(5):
    print(i, end = ' ')
else:
    print("루프 완료 \n")

count = 0
while count < 5:
    print(count, end = ' ')
    count += 1
else:
    print('루프 완료')

# break 문을 사용하여 루프를 중단하면 else절은 실행안됨.
for i in range(5):
    if i == 3:
        print(f'{i}에서 루프를 중단합니다')
        break
else: # else문 break로 실행안됨.
    print('루프 정상완료')

# 출력 : 3에서 루프를 중단합니다
```

#### zip() 함수
- 여러 개의 이터러블(리스트, 튜플 등)을 인자로 받아, 동일한 인덱스에 있는 요소들을 하나의 튜플로 묶어서 이터레이터를 반환
- 데이터를 병렬로 처리할 때 유용

- zip 함수는 이터레이터를 반환하고 이 이터레이터를 단 한번만 소비할 수 있음.
list 함수를 사용하여 이터레이터의 모든 원소를 한 번만 소비해 버리면, 그 결과는
빈 이터레이터가 되기 때문에 리스트로 닷 ㅣ저장한 후 사용해야 됨.

```python
list1 = [1,2,3,4]
list2 = ['a','b','c','d']
zipped = zip(list1, list2)
print(list(zipped))

# Q. 두 개의 리스트(이름과 성)를 사용하여 각 사람의 전체 이름을 생성하세요.
first_name = ['John','Jane','Corey','Travis']
last_name = ['Doe', 'Smith', 'Turner', 'Williams']

full_name = [first + ' ' + last for first, last in zip(first_name, last_name)]
         #      텍스트 작성정리       반복문             zip합수
print(full_name)


# Q. 학생들의 수학과 영어 점수를 각각의 리스트로 받아 평균 점수를 계산하세요.
math_scores = [88,92,79,93,85]
english_scores = [90,91,85,88,86]

avg_scores = [(math+eng) / 2 for math, eng in zip(math_scores, english_scores)]
print(avg_scores)


# Q. 상품 이름과 가격이 각각 다른 리스트에 저장되어 있을 때, 상품명과 가격을 튜플로 묶어 출력하세요.
products = ['apple', 'banana','cherry']
price = [1.2, 0.5, 2.5]

for product, price in zip(products, price):
    print(f'상품:{product}, 가격:${price}')


```

#### 파이썬의 random 모듈
난수를 생성하기 위한 다양한 함수를 제공
- 기본 난수 생성:
  - random.random(): 0과 1 사이의 무작위 실수를 반환.
- 범위 지정 난수 생성:
  - random.randint(a, b): a와 b 사이의 무작위 정수를 반환(b 포함).
  - random.randrange(start, stop[, step]): 지정된 범위에서 무작위 정수를 반환. step을 설정하여 수의 간격을 조정할 수 있다.
- 시퀀스용 함수:
  - random.choice(sequence): 주어진 시퀀스(리스트, 튜플 등)에서 무작위로 아이템을 선택하여 반환.
  - random.shuffle(x[, random]): 리스트의 항목들을 무작위로 섞는다. random은 섞기를 위한 함수를 커스텀할 때 사용.
  - random.sample(population, k): 주어진 시퀀스에서 무작위로 k개의 항목을 뽑아 새 리스트로 반환.
- 분포 기반 난수 생성:
  - random.uniform(a, b): a와 b 사이의 무작위 실수를 균등 분포로 반환.
  - random.gauss(mu, sigma): 평균 mu와 표준편차 sigma를 가지는 정규 분포에서 난수를 반환.

```python
import random
# 0과 1사이의 실수 생성
print(random.random())
#1에서 10사이의 정수 반환 (10포함)
print(random.randint(1,10))
#리스트에서 무작위로 하나의 원소 선택
items = [1,2,3,4,5]
print(random.choice(items))
#리스트의 원소를 무작위로 섞기
random.shuffle(items)
#정규 분포 난수 생성
mean = 0
std_dev = 1
print(random.gauss(mean, std_dev))
```

표현식(Expression)
- 하나 이상의 값, 변수, 연산자, 함수 호출 등을 조합하여 평가(evaluate)될 때 값을 생성하거나 반환하는 코드
- 표현식은 계산되어 값이 되는 모든 것을 의미

표현식의 예시:
- 수치 계산: 2 + 3는 5라는 값을 반환
- 문자열 조합: "Hello " + "World"는 "Hello World"라는 문자열을 생성
- 논리 연산: x > 5는 x의 값에 따라 True 또는 False를 반환
- 함수 호출: sum([1, 2, 3])은 리스트의 합인 6을 반환
- 조건식: x if x > 0 else -x는 x가 양수면 x를, 음수면 -x를 반환

```python
# Task10_0430. 아래 사항을 반영하여 커피 자판기 프로그램을 작성하세요.
# - 시나리오 : 자판기 커피 재고 5잔, 커피 1잔 가격 300원, 재고 범위내에서 300원 이상 돈을 넣으면 거스름돈과 커피를 주고 그렇지 않으면 반환하며 재고가 소진되면 안내멘트 출력
# - 각 Case별 멘트 출력은 상황에 맞게 창작
# - while, if ~ elif ~ else 제어문을 사용하여 작성

def coffee_machine(money):
    price = 300 # 커피값
    count = money // 300 # 커피개수
    change = 0 # 거스름돈
    coffee = 5 # 커피재고

    for i in range(count):
        if money > price and coffee > 0:
            change = money - (i+1) * price
            coffee -= 1
        else :
            print('커피재고 없음')
    print("커피 = %d잔, 잔돈 = %d원" %(i+1, change))

my_money = int(input('투입 금액 :'))
coffee_machine(my_money)


# teach.Task10_0430.
coffee_cnt = 5

while True:
    money = int(input('금액 투입 : '))
    if money == 300:
        print('1잔 나옴')
        coffee_cnt -= 1
    elif money > 300:
        want = int(input('잔수 입력 : '))
        money_want = want * 300
        if money < money_want:
            print('금액 부족')
        else:
            print(f'{want}잔 나옴, 잔돈 {money - money_want}원 받아가세요.')
        coffee_cnt -= want
    else:
        print('금액 부족, 투입금액 반환 (%d원)'%(money))
    if coffee_cnt == 0:
        print('커피 재고 없음')
        break
```
