제어문(control statement)은 프로그램의 흐름을 제어하는데 사용

#### 조건문 (Conditional Statements):
- if: 특정 조건이 참(True)일 때 코드 블록을 실행
    - 조건은 비교 연산자(==, !=, <, >, <=, >=)와 논리 연산자(and, or, not)를 사용하여 표현할 수 있다
- elif: 이전의 if 또는 elif 조건이 거짓(False)일 때 다른 조건을 검사
    - elif는 'else if'의 축약형으로, 이전 if 또는 elif 문의 조건이 거짓(False)일 때 다른 조건을 검사
    - 여러 elif 문을 사용하여 다양한 조건을 검사할 수 있다
- else:
    - 이전의 if 또는 elif 문의 조건이 모두 거짓일 때 실행할 코드 블록을 지정

```python
x = 10
if x > 5:
    print("X는 5보다 크다")
elif x == 5:
    print("X는 5이다")
else:
    print("X는 5보다 작다")


x = 10
if x >= 5:
    print('x는 5보다 크거나 같다')
else:
    print('x는 5보다 작다.')

x = 10
y = 5

if x > 5:
    print("x는 5보다 크다")
if y < 10:
    print("y는 10보다 작다")
if x > y:
    print('x는 y보다 크다')


if x > 5:
    if y >= 3:
        print('x는 5보다 크고 y는 3보다 크거나 같다')
    else:
        print('x는 5보다 크고 y는 3보다 작다')
else:
    print('x는 5보다 작거나 같다')


```

```python
# Q. 이중 조건문을 사용하여 코드를 작성하세요. (조건 내용은 자유. 단, if 문 if ~elif ~else 중첩하여 사용)

score = int(input('점수입력:'))
if score > 70 :
    if score >= 90:
        if score >= 95:
            print('A+')
        else:
            print('A')
    elif score >= 80:
        if score >= 85:
            print('B+')
        else:
            print('B')
    else:
        if score >= 75:
            print('C+')
        else:
            print('C')
else:
    print("낙제")
```

#### 반복문 (Loop Statements):
- for: 시퀀스(리스트, 튜플, 문자열 등)의 아이템을 순회하면서 코드 블록을 실행
- while: 특정 조건이 참인 동안 코드 블록을 반복하여 실행

```python
# for 루프는 시퀀스(리스트, 튜플, 문자열 등)나 반복 가능한 객체(iterable)의 요소를 순회

for i in range(5):
    print(i,end= ' ')
print('\n')
count = 0
while count < 5:
    print(count, end = ' ')
    count += 1
print('\n')
# 리스트를 순회하는 예
for i in [1,2,3,4,5]:
    print(i, end= '\n')

# 문자열을 순회하는 예
for char in 'hello':
    print(char, end =',')
print('\n')

# while 루프는 주어진 조건이 True 동안 코드 블록을 반복
# 조건이 거짓이 되면 루프 종료.
count = 0
sum = 0
while count < 5:
    sum += count
    print(sum, end = ' ')
    count += 1
print('\n')
# Q. d에서 반복문을 사용하여 97 98 99 을 출력하세요.
d = {'a':97, 'b':98, 'c':99}

for key in d:
    print(d[key], end = ' ')
```
